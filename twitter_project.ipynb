```python
#Give me the Z-test O/R T-test over High, low, and close columns and see if the null hypothesis gets rejected or accepted

import pandas as pd
from statsmodels.stats import weightstats as stests
from scipy import stats

df = pd.read_excel("C:/Users/chand/Downloads/TWTR (2).xlsx")

test_value = 30
alpha = 0.05

def perform_ztest(column, test_value):
    ztest, pval = stests.ztest(column, value=test_value)
    return ztest, pval

def perform_ttest(column, test_value):
    ttest_statistic, pval = stats.ttest_1samp(column, test_value)
    return ttest_statistic, pval

def print_results(test_name, column_name, test_statistic, pval):
    print(f"{test_name} for column '{column_name}':")
    print(f"  Test Statistic: {test_statistic}")
    print(f"  P-Value: {pval}")
    if pval < alpha:
        print("  Reject the null hypothesis")
    else:
        print("  Accept the null hypothesis")
    print()

specific_columns = ['High', 'Low', 'Close']
  
for column_name in specific_columns:
    column = df[column_name].dropna()  


    
    # Perform Z-test
    ztest_statistic, ztest_pval = perform_ztest(column, test_value)
    print_results("Z-Test", column_name, ztest_statistic, ztest_pval)
    
    # Perform T-test
    ttest_statistic, ttest_pval = perform_ttest(column, test_value)
    print_results("T-Test", column_name, ttest_statistic, ttest_pval)
```

```python
#Check if the data is dependent or independent by using the chi-square method

df = pd.read_excel("C:/Users/chand/Downloads/TWTR (2).xlsx")

# Creating  categorical columns from continuous data

# Example 1: Categorize 'Close' prices into 'Low', 'Medium', 'High'
close_bins = [0, 40, 45, float('inf')]  # Define bins based on your data
close_labels = ['Low', 'Medium', 'High']
df['Close_Category'] = pd.cut(df['Close'], bins=close_bins, labels=close_labels)

# Example 2: Binarize 'Volume' into 'High Volume' or 'Low Volume'
volume_median = df['Volume'].median()
df['Volume_Category'] = df['Volume'].apply(lambda x: 'High Volume' if x > volume_median else 'Low Volume')

# Create a contingency table
contingency_table = pd.crosstab(df['Close_Category'], df['Volume_Category'])

# Perform Chi-square test of independence
chi2_stat, p_val, dof, expected = stats.chi2_contingency(contingency_table)

# Print results
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-Value: {p_val}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies Table:\n{expected}")

# Interpret the result
alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis: The variables are dependent.")
else:
    print("Accept the null hypothesis: The variables are independent.")
```

```python

import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import io
import base64
from dash import dcc, html
from jupyter_dash import JupyterDash
from dash.dependencies import Input, Output

# Load data
df = pd.read_excel("C:/Users/chand/Downloads/TWTR (2).xlsx")

# Create the Candlestick Chart
candlestick_chart = go.Figure(data=[go.Candlestick(
    x=df['Date'],
    open=df['Open'],
    high=df['High'],
    low=df['Low'],
    close=df['Close']
)])
candlestick_chart.update_layout(
    title='Twitter Stock Prices Over the Years',
    yaxis_title='Stock Price',
    xaxis_title='Date',
    xaxis_rangeslider_visible=True
)

# Create the Bar Chart
bar_chart = px.bar(df, x='Date', y='High', color='Close', title='Comparison of Close vs High Prices')
bar_chart.update_layout(
    xaxis_title='Date',
    yaxis_title='Stock Price'
)

# Create Word Cloud
insights_text = """
Twitter stock prices show varying trends over time. 
Recent trends indicate periods of high volatility, 
with significant fluctuations observed. 
The long-term trend shows an overall growth, 
with periods of significant highs and lows. 
Comparing close and high prices reveals patterns 
of market reactions and investor sentiment.
"""
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate(insights_text)

# Save the Word Cloud to a BytesIO object
wordcloud_img = io.BytesIO()
wordcloud.to_image().save(wordcloud_img, format='PNG')
wordcloud_img.seek(0)
wordcloud_img_b64 = base64.b64encode(wordcloud_img.getvalue()).decode()

# Initialize the Dash app
app = JupyterDash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("Twitter Stock Analysis Dashboard"),

    html.Div([
        dcc.Graph(figure=candlestick_chart)
    ], style={'width': '48%', 'display': 'inline-block'}),

    html.Div([
        dcc.Graph(figure=bar_chart)
    ], style={'width': '48%', 'display': 'inline-block'}),

    html.Div([
        html.H3("Insights"),
        html.Img(src=f'data:image/png;base64,{wordcloud_img_b64}')
    ])
])

# Run the app in Jupyter Notebook
app.run_server(mode='inline')
```
